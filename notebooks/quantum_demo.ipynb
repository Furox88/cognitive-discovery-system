# ⚛️ Quantum Lab: Circuit Simulation

Explore quantum mechanics using the `cds.quantum` module. This notebook shows how to create Bell states and detect entanglement.

```python
from cds.quantum import QuantumCircuit, hadamard, cnot, is_entangled, simulate

# Create a Bell State (|00> + |11>) / sqrt(2)
circuit = QuantumCircuit()
circuit.add(hadamard())
circuit.add(cnot(control=0, target=1))

result_reg = circuit.run()
print(f"Is entangled? {is_entangled(result_reg)}")
print(f"Probabilities: {result_reg.probabilities()}")
```

## Large Scale Sampling

Using our optimized O(1) sampling algorithm to collect 10,000 shots.

```python
counts = simulate(circuit, shots=10000)
print(f"Measurement counts: {counts}")
```
