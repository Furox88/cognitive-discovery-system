# CDS plotting notebook

Requires `pip install cognitive-discovery-system[plot]`.
Companion to `examples/plot_demo.py` and the Cookbook matplotlib section.


```python
import math

from cds.plot import plot_acf, plot_series, plot_spectrum, plot_waveform
from cds.signals import power_spectrum
```

```python
y = [math.sin(2 * math.pi * i / 40) for i in range(100)]
fig = plot_series(y, title="Sine wave")
fig
```

```python
n, fs = 128, 128.0
signal = [math.sin(2 * math.pi * 8 * t / fs) for t in range(n)]
plot_waveform(signal, sample_rate=fs, title="8 Hz tone")
```

```python
power = power_spectrum(signal)
half = [math.sqrt(max(0.0, p)) for p in power[: n // 2]]
plot_spectrum(half, sample_rate=fs / 2, title="Spectrum")
```

```python
plot_acf(y, max_lag=20, title="ACF")
```
